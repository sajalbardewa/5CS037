```python
# 1) Numeric columns stored as text
print(hdi_2022_df.dtypes)               # inspect types
numeric_as_text = [
    col for col in hdi_2022_df.columns
    if hdi_2022_df[col].dtype == "object"
]
print("Object columns:", numeric_as_text)
print(hdi_2022_df[numeric_as_text].head())

# 2) Inconsistent or misspelled country names
print(hdi_2022_df["country"].value_counts().sort_index())

# 3) Duplicate rows
dup_mask = hdi_2022_df.duplicated()
print("Number of duplicate rows:", dup_mask.sum())
dup_rows = hdi_2022_df[dup_mask]
print(dup_rows)

# 4) Special characters representing missing data (e.g. '–')
special_mask = hdi_2022_df.apply(
    lambda col: col.astype(str).str.contains("–")
)
special_counts = special_mask.sum()
print("Special-character '–' counts per column:")
print(special_counts)
```

```python
# 1) Convert data types
hdi_2022_clean = hdi_2022_df.copy()

for col in hdi_2022_clean.columns:
    if hdi_2022_clean[col].dtype == "object" and col not in ["iso3", "country"]:
        hdi_2022_clean[col] = (
            hdi_2022_clean[col]
            .replace("–", pd.NA)          # treat special char as missing
            .astype(str)
            .str.replace(",", "", regex=False)
        )
        hdi_2022_clean[col] = pd.to_numeric(hdi_2022_clean[col], errors="coerce")

# 2) Remove duplicate rows
hdi_2022_clean = hdi_2022_clean.drop_duplicates()

# 3) Handle missing values
#    - For essential identifiers (iso3, country, year, hdi), drop rows with missing values
hdi_2022_clean = hdi_2022_clean.dropna(subset=["iso3", "country", "year", "hdi"])

#    - For remaining numeric columns, impute with column median (keeps all countries for 2022)
numeric_cols = hdi_2022_clean.select_dtypes(include="number").columns
for col in numeric_cols:
    if hdi_2022_clean[col].isna().any():
        median_val = hdi_2022_clean[col].median()
        hdi_2022_clean[col] = hdi_2022_clean[col].fillna(median_val)
```

```python
def classify_hdi(h):
    if h < 0.550:
        return "Low"
    elif h < 0.700:
        return "Medium"
    elif h < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_clean["HDI Category"] = hdi_2022_clean["hdi"].apply(classify_hdi)
```

### 3. Outlier Detection:

```python
df = pd.read_csv("HDI_problem1B.csv")

# Confirm the column exists
print("Columns:", df.columns.tolist())

# Use the exact names as shown in the print above
df_sub = df[["hdi", "gross_inc_percap"]].dropna()

def iqr_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

out_hdi = iqr_outliers(df_sub["hdi"])
out_gni = iqr_outliers(df_sub["gross_inc_percap"])

df_sub["is_outlier_hdi"] = out_hdi
df_sub["is_outlier_gni"] = out_gni
df_sub["is_outlier_any"] = out_hdi | out_gni
```

```python


# 3. Keep only needed columns and drop missing values
df_sub = df[["hdi", "gross_inc_percap"]].dropna()

# 4. IQR-based outlier detection function
def iqr_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

# 5. Detect outliers for HDI and GNI per Capita
out_hdi = iqr_outliers(df_sub["hdi"])
out_gni = iqr_outliers(df_sub["gross_inc_percap"])

# 6. Combined outlier flag
df_sub["is_outlier"] = out_hdi | out_gni

# 7. Scatter plot: GNI per Capita vs HDI with outliers highlighted
plt.figure(figsize=(8, 5))

# Regular points
mask_in = ~df_sub["is_outlier"]
plt.scatter(
    df_sub.loc[mask_in, "gross_inc_percap"],
    df_sub.loc[mask_in, "hdi"],
    alpha=0.5,
    color="steelblue",
    edgecolor="k",
    linewidth=0.2,
    label="Regular"
)

# Outliers
mask_out = df_sub["is_outlier"]
plt.scatter(
    df_sub.loc[mask_out, "gross_inc_percap"],
    df_sub.loc[mask_out, "hdi"],
    alpha=0.9,
    color="red",
    edgecolor="k",
    linewidth=0.3,
    label="Outliers"
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("GNI per Capita vs HDI with Outliers")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python

# Keep only needed columns and drop rows with missing values
cols = ["hdi", "gender_development", "life_expectancy"]
sub = df[cols].dropna()

# Helper function to plot scatter + trendline
def scatter_with_trend(x, y, x_label, y_label, title):
    plt.figure(figsize=(7, 5))

    # Scatter
    plt.scatter(x, y, alpha=0.5, edgecolor="k", linewidth=0.2)

    # Trendline (linear regression using polyfit)
    coeffs = np.polyfit(x, y, 1)
    x_line = np.linspace(x.min(), x.max(), 200)
    y_line = np.polyval(coeffs, x_line)
    plt.plot(x_line, y_line, color="red", linewidth=2, label="Trendline")

    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# 1. Gender Development Index vs HDI
scatter_with_trend(
    sub["gender_development"],
    sub["hdi"],
    "Gender Development Index",
    "HDI",
    "HDI vs Gender Development Index"
)

# 2. Life Expectancy vs HDI
scatter_with_trend(
    sub["life_expectancy"],
    sub["hdi"],
    "Life Expectancy",
    "HDI",
    "HDI vs Life Expectancy"
)
```

## 7. Outlier Detection:

```python
# Column names
hdi_col = "hdi"
gni_col = "gross_inc_percap"

def iqr_outliers(df, col):
    """Return rows that are outliers in df[col] using 1.5 * IQR rule."""
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return df[(df[col] < lower) | (df[col] > upper)], (q1, q3, lower, upper)
```

```python
regions = {"South Asia": south, "Middle East": me}
cols = {"HDI": hdi_col, "GNI per Capita": gni_col}

for region_name, df_region in regions.items():
    for metric_name, col in cols.items():
        outliers, stats = iqr_outliers(df_region, col)
        q1, q3, lower, upper = stats
        print(f"{region_name} - {metric_name} outliers:")
        print(outliers[["country", "year", col]])
        print(f"{metric_name} Q1={q1:.3f}, Q3={q3:.3f}, lower={lower:.3f}, upper={upper:.3f}")
        print()
```

```python
# Compute bounds and outlier flags
south_hdi_lower, south_hdi_upper = iqr_bounds(south[hdi_col])
south_gni_lower, south_gni_upper = iqr_bounds(south[gni_col])

me_hdi_lower, me_hdi_upper = iqr_bounds(me[hdi_col])
me_gni_lower, me_gni_upper = iqr_bounds(me[gni_col])

south["hdi_outlier"] = (south[hdi_col] < south_hdi_lower) | (south[hdi_col] > south_hdi_upper)
south["gni_outlier"] = (south[gni_col] < south_gni_lower) | (south[gni_col] > south_gni_upper)

me["hdi_outlier"] = (me[hdi_col] < me_hdi_lower) | (me[hdi_col] > me_hdi_upper)
me["gni_outlier"] = (me[gni_col] < me_gni_lower) | (me[gni_col] > me_gni_upper)

# 1. South Asia: HDI vs GNI per capita
plt.figure(figsize=(6, 4))
# Non-outliers
mask = ~south["hdi_outlier"] & ~south["gni_outlier"]
plt.scatter(south.loc[mask, gni_col],
            south.loc[mask, hdi_col],
            color="steelblue", alpha=0.6, label="Non-outliers")
# Outliers
mask_out = south["hdi_outlier"] | south["gni_outlier"]
plt.scatter(south.loc[mask_out, gni_col],
            south.loc[mask_out, hdi_col],
            color="red", alpha=0.9, label="Outliers")

plt.title("South Asia: HDI vs GNI per Capita")
plt.xlabel("GNI per Capita Index")
plt.ylabel("HDI")
plt.legend()
plt.tight_layout()
plt.show()

# 2. Middle East: HDI vs GNI per capita
plt.figure(figsize=(6, 4))
mask = ~me["hdi_outlier"] & ~me["gni_outlier"]
plt.scatter(me.loc[mask, gni_col],
            me.loc[mask, hdi_col],
            color="steelblue", alpha=0.6, label="Non-outliers")

mask_out = me["hdi_outlier"] | me["gni_outlier"]
plt.scatter(me.loc[mask_out, gni_col],
            me.loc[mask_out, hdi_col],
            color="red", alpha=0.9, label="Outliers")

plt.title("Middle East: HDI vs GNI per Capita")
plt.xlabel("GNI per Capita Index")
plt.ylabel("HDI")
plt.legend()
plt.tight_layout()
plt.show()
```
